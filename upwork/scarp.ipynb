```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

books = []

for i in range(1, 5):
    # Make a request to a webpage
    url = f"https://books.toscrape.com/catalogue/page-{i}.html"
    response = requests.get(url)
    
    # Extract HTML content from the response
    html_content = response.content
    
    # Create a BeautifulSoup object
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # Find the <ol> tag containing all the book articles
    ol = soup.find('ol', class_='row')
    
    # Find all the <article> tags representing individual books
    articles = ol.find_all('article', class_='product_pod')
    
    # Loop through each article to extract book information
    for article in articles:
        image = article.find('img')
        title = image.attrs['alt']
        star = article.find('p')['class'][1]
        price = article.find('p', class_='price_color').text
        price = float(price[1:])  # Remove the currency symbol and convert to float
        books.append([title, price, star])


df = pd.DataFrame(books,columns=['Title','Price','Star Rating'])
df.to_excel('books.xlsx')
```
